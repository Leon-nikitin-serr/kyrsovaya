```python
img_path = files[0]
img = read_img_path(img_path)  # → Tensor CHW, значения 0–255

# Подготовка массива 
# CHW → HWC, нормализация 0–1
img_hwc = np.transpose(img.numpy(), (1, 2, 0)) / 255.0
print("Форма HWC:", img_hwc.shape)  # (3333, 5000, 3)
```

```python
# Берём красный и зелёный каналы
R = img_hwc[:, :, 0]  
G = img_hwc[:, :, 1]
print("R shape:", R.shape, "| max R:", R.max())
R = R / R.max()
G = G / G.max()
print("R shape:", R.shape, "| max R:", R.max())
```

Создаём полосочный фильтр K

```python
h, w = R.shape  # размеры изображения

stripe_width = 70  # толщина полос

x = np.arange(w)
y = np.arange(h)

X, Y = np.meshgrid(x, y)

K = ((Y // stripe_width) % 2).astype(float)
```

```python
print("Доля R:", K.mean())
```

Генерация измерения P

```python
P = K * R ** 0.7 + (1 - K) * G ** 2
```

```python
# Визуализация
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

axes[0, 0].set_title("R (красный канал)")
axes[0, 0].imshow(R**0.7, cmap='gray')
axes[0, 0].axis("off")

axes[0, 1].set_title("G (зелёный канал)")
axes[0, 1].imshow(G**2, cmap='gray')
axes[0, 1].axis("off")

axes[1, 0].set_title("K (полосатый фильтр)")
axes[1, 0].imshow(K, cmap='gray')
axes[1, 0].axis("off")

axes[1, 1].set_title("P (синтетическое измерение)")
axes[1, 1].imshow(P, cmap='gray')
axes[1, 1].axis("off")

plt.tight_layout()
plt.show()
```

```python
#  Добавление шума
# σ=10 при диапазоне 0–255, но у нас 0–1 → σ = 10/255 ≈ 0.039
sigma = 10 / 255.0
noise = np.random.normal(0, sigma, P.shape)
P_noisy = np.clip(P + noise, 0, 1)  # clip чтобы не выйти за [0,1]

plt.figure(figsize=(8, 6))
plt.title(f"P с гауссовским шумом (σ={sigma:.3f})")
plt.imshow(P_noisy, vmin=0, vmax=1)
plt.axis("off")
plt.show()
```

```python
import torch
import torch.nn as nn

# вход сети
input_img = torch.tensor(P_noisy, dtype=torch.float32).unsqueeze(0).unsqueeze(0)

# target (оригинал)
target = np.stack([R, G], axis=0)
target = torch.tensor(target, dtype=torch.float32).unsqueeze(0)

print("Input shape:", input_img.shape)
print("Target shape:", target.shape)
```

```python
class StripeNet(nn.Module):

    def __init__(self):
        super().__init__()

        self.net = nn.Sequential(

            nn.Conv2d(1, 32, 3, padding=1),
            nn.ReLU(),

            nn.Conv2d(32, 64, 3, padding=1),
            nn.ReLU(),

            nn.Conv2d(64, 32, 3, padding=1),
            nn.ReLU(),

            nn.Conv2d(32, 2, 3, padding=1)
        )

    def forward(self, x):
        return self.net(x)
```

```python
model = StripeNet()

criterion = nn.MSELoss()

optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
```

```python
import cv2

small = cv2.resize(P_noisy, (512,512))
R_small = cv2.resize(R, (512,512))
G_small = cv2.resize(G, (512,512))

input_img = torch.tensor(small).float().unsqueeze(0).unsqueeze(0)
target = torch.tensor(np.stack([R_small,G_small])).float().unsqueeze(0)
```

```python
for epoch in range(300):

    optimizer.zero_grad()

    output = model(input_img)

    loss = criterion(output, target)

    loss.backward()

    optimizer.step()

    if epoch % 50 == 0:
        print("epoch:", epoch, "loss:", loss.item())
```

```python
with torch.no_grad():
    pred = model(input_img)

pred = pred.squeeze().numpy()

R_pred = pred[0]
G_pred = pred[1]
```

```python
fig, ax = plt.subplots(2,2, figsize=(10,8))

ax[0,0].set_title("Original R")
ax[0,0].imshow(R_small, cmap='gray')
ax[0,0].axis("off")

ax[0,1].set_title("Predicted R")
ax[0,1].imshow(R_pred, cmap='gray')
ax[0,1].axis("off")

ax[1,0].set_title("Original G")
ax[1,0].imshow(G_small, cmap='gray')
ax[1,0].axis("off")

ax[1,1].set_title("Predicted G")
ax[1,1].imshow(G_pred, cmap='gray')
ax[1,1].axis("off")

plt.show()
```

```python
from skimage.metrics import peak_signal_noise_ratio

psnr_R = peak_signal_noise_ratio(R_small, R_pred)
psnr_G = peak_signal_noise_ratio(G_small, G_pred)

print("PSNR R:", psnr_R)
print("PSNR G:", psnr_G)
```

```python
from skimage.metrics import structural_similarity

ssim_R = structural_similarity(R_small, R_pred, data_range=1)
ssim_G = structural_similarity(G_small, G_pred, data_range=1)

print("SSIM R:", ssim_R)
print("SSIM G:", ssim_G)
```
