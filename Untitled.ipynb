```python
import os
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
from olimp.dataset import read_img_path

# Базовый кэш
cache_dir = r"C:\Users\NikLeo\.cache\pyolimp\OLIMP\OLIMP"

# Категория и подкатегория
category = "urban scenes"
subcategory = "architecture"
category_folder = Path(cache_dir) / category / subcategory

# Список файлов
files = list(category_folder.iterdir())
print("Файлы в категории:", files[:5])

# Берём первый файл
img_path = files[0]
img = read_img_path(img_path)  # теперь path — Path, как требует OLIMP

# Конвертируем CHW → HWC для matplotlib
img = np.transpose(img, (1, 2, 0))
img = img/255
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
# img = read_img_path("путь_к_изображению")  # используем OLIMP
# Для корректного отображения matplotlib нужно поменять оси CHW → HWC
img = np.transpose(img, (1, 2, 0))
print("Форма изображения после транспозиции:", img.shape)
```

```python
# Берём красный и зелёный каналы
R = img[:, :, 0]  # Красный канал, форма HxW
G = img[:, :, 1]  # Зеленый канал, форма HxW

print("Форма R:", R.shape, "Форма G:", G.shape)
```

```python
# Если R и G — PyTorch Tensor
R = R.numpy()
G = G.numpy()
print(R.max())
```

Создаём полосочный фильтр K

```python
h, w = R.shape  # размеры изображения

stripe_width = 50  # толщина полос

x = np.arange(w)
y = np.arange(h)

X, Y = np.meshgrid(x, y)

K = ((X // stripe_width) % 2).astype(float)
```

```python
print("Доля R:", K.mean())
```

Генерация измерения P

```python
P = K * R + (1 - K) * G
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.title("R")
plt.imshow(R)
plt.axis("off")

plt.subplot(2,2,2)
plt.title("G")
plt.imshow(G)
plt.axis("off")

plt.subplot(2,2,3)
plt.title("K")
plt.imshow(K)
plt.axis("off")

plt.subplot(2,2,4)
plt.title("P")
plt.imshow(P)
plt.axis("off")

plt.show()
```

```python
# Добавляем гауссовский шум
noise = np.random.normal(0, 10, P.shape)  # среднее=0, σ=10
P_noisy = P + noise

# Показываем P с шумом
plt.figure(figsize=(6, 6))
plt.title("Синтетическое измерение с шумом (P_noisy)")
plt.imshow(P_noisy)
plt.axis("off")
plt.show()
```
